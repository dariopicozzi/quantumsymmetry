# Welcome: the binary-tree ansatz

This second tutorial series is a companion to the manuscript

> *Picozzi, D. (manuscript). A universal hardware-efficient variational ansatz with exact diagonal metric for real- and imaginary-time evolution and Haar sampling.*

It introduces the `MinimalCircuit` ansatz that `quantumsymmetry` exposes alongside the symmetry-adapted encodings of the [first tutorial series](01_welcome.ipynb), and shows how to reproduce, noiselessly and in a few minutes each, the simplest version of every numerical experiment in the paper.

## The three main ideas, informally

A variational circuit prepares a parametrised quantum state $|\psi(\boldsymbol\lambda)\rangle$ whose parameters we tune to solve a problem (find a ground state, follow a time evolution, optimise a cost). The paper builds **one** ansatz with three properties that usually do not come together:

**1. A universal binary tree.** Imagine the $2^n$ computational-basis states as the leaves of a binary tree. Starting from the root, each internal node carries one rotation angle $\theta_i$ that splits the amplitude between its two sub-trees (a $\cos\theta_i$ / $\sin\theta_i$ weighting). The amplitude of any basis state is then simply the product of the factors along the path from the root to its leaf. Walking the whole tree reproduces the standard hardware-efficient *CNOT staircase*, so the circuit is both universal (it can reach any state) and shallow.

**2. A diagonal metric, for free.** The natural notion of distance between nearby quantum states is the Fubini-Study metric. For a generic ansatz it is a dense matrix that must be estimated with many extra circuits and then inverted. For the binary tree it is **diagonal**, with entries you can read straight off the tree, so the geometry-aware algorithms (quantum natural gradient, imaginary- and real-time evolution) cost no extra circuits at all.

**3. Pruning for sparse targets.** If the state you want lives in a small subspace -- a fixed particle-number or symmetry sector, a fixed-Hamming-weight feasible set -- then most of the tree is redundant. Marking the *active leaves* lets the compiler prune the circuit, with a CNOT count that grows **linearly** in the number $k$ of active leaves instead of exponentially in $n$.

Everything in the series is driven by a single object, `MinimalCircuit`. Let us build the smallest possible example: the four basis states of the $H_2$ molecule's $(1\uparrow, 1\downarrow)$ particle-number sector.

```python
from quantumsymmetry import MinimalCircuit

# Two spatial orbitals (4 qubits), one spin-up and one spin-down electron.
mc = MinimalCircuit.from_particle_number(num_spatial_orbitals=2,
                                         num_particles=(1, 1))

print('qubits          :', mc.num_qubits)
print('active leaves   :', mc.support)
print('free parameters :', mc.num_parameters)
print('CNOTs           :', mc.circuit.count_ops().get('cx', 0))
```

Four active leaves on four qubits, prepared by a handful of gates. The next notebook opens up this object -- the tree, the pruned circuit and the diagonal metric -- before we put it to work on the paper's applications.
